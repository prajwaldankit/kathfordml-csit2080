```python
year_country_counts = df.groupby(['release_year', 'country']).size()
print(year_country_counts.head())

# Find the top country for each year
top_country_per_year = year_country_counts.groupby('release_year').idxmax()
print(top_country_per_year.head())

result = pd.DataFrame(top_country_per_year.tolist(), columns=[
                      'release_year', 'top_country'])
result['num_releases'] = year_country_counts.loc[top_country_per_year].values
result
```
