```python
def decorator_function(func):
    def wrapper(*args, **kwargs):
        print("before running")
        return func(*args, **kwargs)
    
    return wrapper

def time_taken_calculator(func):
    def wrapper(*args, **kwargs):
        import datetime
        before_time = int(datetime.datetime.now().timestamp())
        result = func(*args, **kwargs)
        after_time = int(datetime.datetime.now().timestamp())
        print(f"timetaken is: {after_time - before_time}")
        return result
        
    
    return wrapper


@time_taken_calculator
def add(a,b):
    import time 
    time.sleep(1)
    print(a + b)

add(1 ,2)
# decorated_test = decorator_function(test)

# decorated_test()
```
