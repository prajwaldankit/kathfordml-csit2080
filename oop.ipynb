```python
class Student:
    def __init__(self, name, roll, age, marks):
        self.name = name
        self.age = age
        self.roll = roll
        self.marks = marks
    
    def add_marks(self,mark):
        if mark < 0:
            raise ValueError
        self.marks.append(mark)

    def get_avg(self):
        return sum(self.marks) / len(self.marks)

    def get_grade(self):
        avg = self.get_avg()
        if avg > 80: 
            return "A"
        
        if avg > 70: 
            return "b"
        
        if avg > 55: 
            return "c"

        return "F"

        

s1 = Student("MIchael Jordan", 23, 60, [63,54,67])
```

```python
s1.get_avg()
s1.add_marks(34)
s1.get_grade()
```
