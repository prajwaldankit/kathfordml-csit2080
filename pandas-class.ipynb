```python
students = {
    "Name": ["Ram", "Sita", "Ram", "Gita", "Ram"],
    "Department": ["CS", "CS", "IT", "IT", "HR"],
    "Marks": [80, 92, 75, 88, 60]
}

df = pd.DataFrame(students)
print(df)
print(df[df["Marks"] > 80])
```

```python
df.groupby("Department")["Marks"].max()
```

```python
print(df.groupby("Department")["Marks"].agg(["mean", "sum", "count"]))
```

```python
print(df["Marks"].agg(["mean", "max"]))
```

```python
print(
    df.groupby(["Department", "Name"])["Marks"].mean()
)
```
