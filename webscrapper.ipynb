- div.card
    - img -> product image
    - div.captions(product details)
        - h4.price > span.price (price of the product)
        - h4 > a.title (product title)
        - p.description (product description)
    - div.ratings
        - p.review-count > span (number of reviews)
        - p > multples spans (num of spans is the star rating)

```python
import requests
from bs4 import BeautifulSoup
```

```python
BASE_URL = "https://webscraper.io/test-sites/e-commerce/allinone"

response = requests.get(BASE_URL)


print("Status:", response.status_code)
print("Content-Type:", response.headers['Content-Type'])
```

```python
print(response.text)
```

```python
soup = BeautifulSoup(response.text, "html.parser")

print(soup)
```

```python
cards = soup.select("div.card")
print(cards)
print(len(cards))
```

```python
card = cards[0]

image = card.select_one("img")
caption = card.select_one("div.caption")
print(caption)
```
